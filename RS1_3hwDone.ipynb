```python
# GENRE = 'Action'
GENRE = 'Comedy'
# решила расчитать ранг фильма, как сумму оценок по фильму, деленную на время :)
```

```python
find_movies_by_genre = movies[movies['genres'].str.find(GENRE) != -1]
find_movies_by_genre.head()
```

```python
movies_by_genre_with_ratings = find_movies_by_genre.join(ratings.set_index('movieId'), on='movieId')
movies_by_genre_with_ratings
```

```python
sum_rating_movies_by_genre = movies_by_genre_with_ratings.groupby(['movieId'])[['rating']].sum().sort_values('movieId').head(20)
sum_rating_movies_by_genre
```

```python
lasttime_movies_by_genre = movies_by_genre_with_ratings.groupby(['movieId'])[['timestamp']].max().sort_values('movieId').head(20)
lasttime_movies_by_genre
```

```python
top_movies_by_genre = sum_rating_movies_by_genre.join(movies.set_index('movieId'), on='movieId')
top_movies_by_genre['timestamp'] = lasttime_movies_by_genre
top_movies_by_genre['rate'] = top_movies_by_genre['rating']/top_movies_by_genre['timestamp']*100000000
top_movies_by_genre.sort_values('rate',ascending=False).head(20)
```
